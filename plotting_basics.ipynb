# Manipulating axillary bits plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

initially just produsing two arrays (lists) or 100 random numbers between 0 and 100

```python
x = np.random.randint(0,100,100)
y = np.random.randint(0,100,100)
```

### Very basic plot

```python
plt.scatter(x,y)
```

### setting the axis ranges

```python
plt.scatter(x,y)
plt.ylim(-100,200)
plt.xlim(0,500)
```

### changing the color of the points

```python
plt.scatter(x,y,color='red')
```

### setting axis lables

```python
plt.scatter(x,y)
plt.ylabel('this is the y lable')
plt.xlabel('this is the x lable')
```

### Adding a legend

```python
plt.scatter(x,y,label='my random numbers')
plt.scatter(x,y*10.0,color='red',label='my random numbers with y timesed by 10')

plt.legend()
```

more detals on how to customise the legend are here: https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html

### Adding a title

```python
plt.scatter(x,y)
plt.title('my lovely plot')
```

### Saving a figure

```python
plt.scatter(x,y)
plt.savefig('my_lovely_plot.png') # note, this will save it in the location you are currently working in - you can also specify the path to wher eyou want to save it
```

#### Saving as a diffreent file type to a png, just change teh file extension. 
This could be useful for exmaple if you wanted to manipulate the possibtions of subplots or change text in a vector graphics packafe liek inkscape

```python
plt.scatter(x,y)
plt.savefig('my_lovely_plot.svg')
```

OR

```python
plt.scatter(x,y)
plt.savefig('my_lovely_plot.pdf')
```

# Using subplots

```python
#plot 1
plt.subplot(1, 2, 1) # This means had a grid of 1 plot down, two plots across, and start with the 1st plot
plt.scatter(x,y)

#plot 2:
plt.subplot(1, 2, 2) # This means had a grid of 1 plot down, two plots across, and now do the 2nd plot
plt.scatter(x,y)
```

As above but 3 by 2

```python
plt.subplot(3, 2, 1) 
plt.scatter(x,y,color = 'black')

plt.subplot(3, 2, 2) 
plt.scatter(x,y,color = 'yellow')

plt.subplot(3, 2, 3) 
plt.scatter(x,y,color = 'red')

plt.subplot(3, 2, 4) 
plt.scatter(x,y,color = 'blue')

plt.subplot(3, 2, 5) 
plt.scatter(x,y,color = 'green')

plt.subplot(3, 2, 6) 
plt.scatter(x,y,color = 'grey')

plt.tight_layout() # Note, this command tells it to space the figures out sensibly.
```

### Changing the figure size

```python
plt.figure(figsize=(10,10)) # size across and down is 10 and 10 (arbitrary units)

plt.subplot(3, 2, 1) 
plt.scatter(x,y,color = 'black')

plt.subplot(3, 2, 2) 
plt.scatter(x,y,color = 'yellow')

plt.subplot(3, 2, 3) 
plt.scatter(x,y,color = 'red')

plt.subplot(3, 2, 4) 
plt.scatter(x,y,color = 'blue')

plt.subplot(3, 2, 5) 
plt.scatter(x,y,color = 'green')

plt.subplot(3, 2, 6) 
plt.scatter(x,y,color = 'grey')

plt.tight_layout() # Note, this command tells it to space the figures out sensibly.
```
